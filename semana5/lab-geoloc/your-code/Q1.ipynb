El primer paso es seleccionar lo datos desde MongoDB, pasarlos a un dataframe, limpiarlos y guardar las seleccion de datos en un archivo json para volver a llevarlos a MongoDB.

```python
# Tu codigo aqui

from pymongo import MongoClient
import pandas as pd
import numpy as np
import json
```

```python
client = MongoClient()
db = client.db_companies
```

```python
db.companies.count()
```

```python
companies_df = pd.DataFrame(list(db.companies.find({"$and" : 
                                                        [{"$or": 
                                                              [{"category_code" : "web"},
                                                                {"category_name" : "software"},
                                                                {"category_name" : "games_video"},
                                                                {"category_name" : "mobile"},
                                                                {"category_name" : "photo_video"},
                                                                {"category_name" : "network_hosting"},
                                                                {"category_name" : "ecommerce"}]},
                                                         {"$and": 
                                                                [{"number_of_employees" : {"$gt" : 0}},
                                                                {"offices" : {"$ne" : []}},
                                                                {"deadpooled_year": None},
                                                                {'total_money_raised':{'$ne':'$0'}},
                                                                {'founded_year':{'$gt':0}},
                                                                {'offices.latitude':{'$ne':None}},
                                                                {'offices.longitude':{'$ne':None}}
                                                                ]}]
                                                   }
                                                  )))
companies_df
```

```python
usage_companies = pd.DataFrame.from_dict(companies_df).sort_values('founded_year',ascending=True)
usage_companies
```

```python
def clasify_company(x):
    if x.founded_year > 2005 and x.number_of_employees < 300:
        return 'Start-up'
    else:
        return 'Company'


def office(x):
    field = {"type": "Point","coordinates": [x[0]['longitude'],x[0]['latitude']]}
    return field

usage_companies["office_loc"] = usage_companies.offices.apply(lambda x: office(x))
usage_companies["company_type"] = usage_companies.apply(clasify_company, axis = 1)
```

```python
len(usage_companies[usage_companies.company_type == "Company"])
```

```python
len(usage_companies[usage_companies.company_type != "Company"])
```

```python
content = []
dict_companies = usage_companies.transpose().to_dict()
for i in range(len(dict_companies)):
    content.append(dict_companies[i])
```
